### This section does a sentiment analysis on 50 of the Chicago Bears most recent tweets.

```python
import pickle
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

with open('Bears_Large.pkl', 'rb') as f:
    chiefs = pickle.load(f)
    
def is_happy(tweet):
    score = analyzer.polarity_scores(tweet.text)
    return score['pos'] > 0.2

#happy_tweets = filter(is_happy, chiefs)
#for tweet in happy_tweets:
    #print(tweet.text)
    
def plottable_sentiments(tweets):
    texts = [tweet.text for tweet in tweets]
    sents = [analyzer.polarity_scores(t) for t in texts]
    pos_x = [s['pos'] for s in sents]
    neu_y = [s['neu'] for s in sents]
    neg_z = [s['neg'] for s in sents]
    return pos_x, neu_y, neg_z
    
chiefs_texts = []
for tweet in chiefs:
    chiefs_texts.append(tweet.text)
    
chiefs_texts = [tweet.text for tweet in chiefs]
chiefs_sentiments = [analyzer.polarity_scores(t) for t in chiefs_texts]
#print(chiefs_sentiments)
    
# To plot 3d scatter plot
# We need 3 lists: x vals, y vals, z vals
pos_x = [sent['pos'] for sent in chiefs_sentiments]
neu_y = [sent['neu'] for sent in chiefs_sentiments]
neg_z = [sent['neg'] for sent in chiefs_sentiments]
# boilerplate
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
rbx, rby, rbz = plottable_sentiments(chiefs)

with open('Bears_Large.pkl', 'rb') as f:
    chiefs_tweets = pickle.load(f)
    
qx, qy, qz = plottable_sentiments(chiefs_tweets)

ax.scatter(rbx, rby, rbz, c='r', marker='o')
ax.scatter(qx, qy, qz, c = 'purple', marker='^')
ax.set_xlabel('Positive')
ax.set_ylabel('Neutral')
ax.set_zlabel('Negative')
plt.show()
```
